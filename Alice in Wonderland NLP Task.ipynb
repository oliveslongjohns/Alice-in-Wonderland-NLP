```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
  
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
sia = SentimentIntensityAnalyzer()
```

## Task 1

```python
with open('alice_in_wonderland.txt',encoding='UTF-8') as file:
    aiw = file.read()

aiw = aiw[aiw.find('CHAPTER I.\nDown the Rabbit-Hole'):aiw.find('THE END')+7]
## Read all past first chapter title until end (including endpoints).

aiw_l = []
roman_num = ['I','II','III','IV','V','VI','VII','VIII','IX','X','XI','XII']
for i in range(0,11):
    aiw_l += [aiw[aiw.find('CHAPTER {}.'.format(roman_num[i])):aiw.find('CHAPTER {}.'.format(roman_num[i+1]))]]
aiw_l += [aiw[aiw.find('CHAPTER XII'):aiw.find('THE END')]]
## Segment corpus into Chapters, based on title.

df = pd.DataFrame()
for i in range(0,12):
    df = pd.concat([df,pd.DataFrame([[i+1,x.lower().replace('—','').replace('-','').replace('_','')] for x in word_tokenize(aiw_l[i]) if x not in '.,;:_?!“”‘’"\'£$%^&*()'])])
df.rename({0:'Chapter',1:'Token'},axis='columns',inplace=True)
## Iterate through each chapter, so we can add chapter number as a label.
## Tokenize on word, whilst dealing with the natural symbols in the text. 
## Some characters (-,_,—) occur attached to tokens, so we remove these from individual strings.

df['num_occurences'] = '1'
kw_count = df.groupby(['Chapter','Token']).count().reset_index()
## Count keywords over all the chapters with a group by aggregation.

kw_count.to_csv('alice_in_wonderland word counts.csv',index=False)
## Save to CSV.
```

## Task 2

### Appearance of Main Characters by Chapter

The first analytical insights we shall derive here is the measuring the appearance of main characters across the chapters. To pose this as a question, "Which main characters feature in what chapters?" is what we shall attempt to answer.

As a prequisite to this, we must locate the characters whose names form n-grams, and ensure they will remain as single tokens after our processing. We'll start by extracting all the most common bigrams from the complete text.

```python
words = [lemmatizer.lemmatize(x.lower()).replace('—','').replace('-','').replace('_','') for x in word_tokenize(aiw) if (x not in '.,;:_?!“”‘’"\'£$%^&*()') and (x.lower() not in stop_words)]
## Word Tokenize across the whole novel, just for finding bigrams.
## We remove stop words using NLTK's prebuilt stop-words list, and lemmatize using NLTK's Lemmatizer.
bi_finder = nltk.collocations.BigramCollocationFinder.from_words(words)
## We use NLTK's Bigram Collocation Finder for identifying bigrams.
```

```python
bi_finder.ngram_fd.most_common(58)
## We sample the 58 most common bigrams arbitrarily, so that we can show each bigram-named character.
```

Quite naturally for a novel, the bigrams are dominated by dialogue patterns (e.g. 'said Alice'), but we do spot 5 interesting character names within the 58 most common bigrams. 'Mock Turtle', 'March Hare' (The March Hare), 'White Rabbit', 'Cheshire Cat' and 'three gardener' (three gardeners). 

For 3 of these cases, White Rabbit, Cheshire Cat and three gardeners, they are not always referred to with their complete title (e.g. just 'rabbit', 'cat' or 'gardeners'). As they are still identifiable by their non-proper noun, we won't make any change here. For Mock Turtle and March Hare, however, we will remove the space in their names, for the purposes of our analysis.

Note that we also checked trigrams and quadgrams, but there were no observations of note, so we've omitted it here.

```python
aiw2 = aiw.replace('\n',' ').replace('March Hare','MarchHare').replace('Mock Turtle','MockTurtle')
## We also must deal with the new lines (\n) in case their name is spread across two lines.
```

We now repeat the same steps as in Task 1, with some additions to enrich the data. We retain word types using NLTK's part-of-speech tagging, as well as lemmatize and remove stop words.

```python
aiw2_l = []
roman_num = ['I','II','III','IV','V','VI','VII','VIII','IX','X','XI','XII']
for i in range(0,11):
    aiw2_l += [aiw2[aiw2.find('CHAPTER {}.'.format(roman_num[i])):aiw2.find('CHAPTER {}.'.format(roman_num[i+1]))]]
aiw2_l += [aiw2[aiw2.find('CHAPTER XII'):aiw2.find('THE END')]]
## Repeat chapter segmentation for modified novel text.

df2 = pd.DataFrame()
for i in range(0,12):
    df2 = pd.concat([df2,pd.DataFrame([[i+1,lemmatizer.lemmatize(x[0].lower()).replace('—','').replace('-','').replace('_','').replace('.',''),x[1]] for x in nltk.pos_tag(word_tokenize(aiw2_l[i])) if (x[0] not in ' .,;:_?!“”‘’"\'£$%^&*()[]') and (x[0].lower() not in stop_words)])])
## Use nltk.pos_tag for classifying word type. This function uses part-of-speech tagging to determine word type.
## It's worth noting that this uses a combination of word definition and grammatical context, but it's not perfect, and we will need to do some cleaning.
df2.rename({0:'Chapter',1:'Token',2:'Word Type'},axis='columns',inplace=True)
df2 = df2[df2['Token']!='']
df2['Num Occurrences'] = '1'
kw_count2 = df2.groupby(['Chapter','Token','Word Type']).count().reset_index()
```

```python
kw_count2[kw_count2['Token']=='alice']
```

A list of all possible word types in POS tagging can be found here: https://www.ling.upenn.edu/courses/Fall_2003/ling001/penn_treebank_pos.html

Things generally look consitent, but one problem is that our word type tagging is sometimes inconsistent. Alice should always be a proper noun, but there are exception cases. To clean further, we will 'impute' the word types for every word to be the most common type tag that word received. For example, we will change every occurence of 'Alice' to be 'NNP', as it is the most common word type. This will make it easier to analyse, and also make our table of counts a unique combination of Chapter and Token.

```python
kw_types = kw_count2[['Token','Word Type','Num Occurrences']].groupby(['Token','Word Type']).sum().reset_index()
## First sum across all chapters, to use all available data for each token.
kw_types['Type Rank'] = kw_types.groupby('Token')['Num Occurrences'].rank(method='first',ascending=False)
## Type Rank refers to the rank of that word type across each partition of tokens. See example beneath.
```

```python
kw_types[kw_types['Token']=='alice']
```

In the case of 'alice', we can see that NNP is the dominant word type and therefore of rank 1. 

```python
kw_types = kw_types[kw_types['Type Rank'] == 1][['Token','Word Type','Num Occurrences']].rename({'Word Type':'Imputed Word Type','Num Occurrences':'Total Occ.'},axis='columns')
## Filter to only type ranks of value 1, and we have a dataframe of each most popular word type for each token.
kw_count2 = kw_count2.merge(kw_types,how='left',on='Token')
## Join to our original table to obtain 'Imputed Word Type in each instance'
kw_count2 = kw_count2.groupby(['Chapter','Token','Imputed Word Type']).agg({'Num Occurrences':sum,'Total Occ.':max}).reset_index()
## Now aggregate on Chapter, Token and Imputed Word Type, summing number of occurences for each.
```

```python
kw_count2[kw_count2['Token']=='alice']
```

We now have a unique record for each Chapter and Token.

Next, we shall attempt to visualise what characters features and how often across each chapter. 

```python
main_char = kw_count2[(kw_count2['Imputed Word Type']=='NNP') & (kw_count2['Total Occ.']>25)]
```

First, we filter our dataframe to proper nouns (NNP) and filter out any tokens that had total count below 25 across the whole corpus.

```python
mc_piv = main_char[['Chapter','Token','Num Occurrences']].pivot_table(index='Token',columns='Chapter',aggfunc=sum,fill_value=0)
## Pivot, to get a 2 dimensional table of keyword counts across Chapter and Character.

pref_names = {'alice':'Alice','rabbit':'Rabbit','cat':'Cheshire Cat','mouse':'Mouse',
              'duchess':'Duchess','caterpillar':'Caterpillar',
              'hatter':'Mad Hatter','dormouse':'Dormouse','marchhare':'March Hare',
              'queen':'Queen of Hearts','king':'King of Hearts',
              'mockturtle':'Mock Turtle','gryphon':'Gryphon'}
mc_piv = mc_piv.reindex(pref_names.keys())
mc_piv.rename(pref_names,inplace=True)
## We reorder the names and also relabel, for neatness of the visualisation.
```

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
im = ax.imshow(X=np.arange(100).reshape((10,10)),cmap = sns.color_palette("light:orange", as_cmap=True))

cax = fig.add_axes([ax.get_position().x1+0.04,ax.get_position().y0,0.04,ax.get_position().height])
cbar = fig.colorbar(im, cax=cax) # Similar to fig.colorbar(im, cax = cax)
cbar.set_ticks([0,25,50,75,99])
cbar.ax.set_yticklabels(["0","10","20","30","40+"])
# Create an axes for colorbar. The position of the axes is calculated based on the position of ax.

chaps = ["I. Down the Rabbit-Hole","II. The Pool of Tears","III. A Caucus-Race and a Long Tale",
         "IV. The Rabbit Sends in a Little Bill","V. Advice from a Caterpillar","VI. Pig and Pepper",
         "VII. A Mad Tea-Party","VIII. The Queen’s Croquet-Ground","IX. The Mock Turtle’s Story",
         "X. The Lobster Quadrille","XI. Who Stole the Tarts?","XII. Alice’s Evidence"]
# Chapter Names as a list.

sns.heatmap(
    mc_piv,
    vmin = 0,
    vmax = 40,
    center = 20,
    cmap = sns.color_palette("light:orange", as_cmap=True),
    square = True,
    linewidths = 0.5,
    annot = True,
    ax=ax,
    cbar=False)
ax.set_title("Count of Character Mentions by Chapter",fontsize=15,pad=15);
ax.set_xlabel("Chapter", size=15, labelpad=20)
ax.set_ylabel("Character", size=15, labelpad=20)
ax.set_xticklabels(chaps, fontsize="medium", rotation=45, ha="right", rotation_mode="anchor");
## Count of Character Mentions by Chapter
```

As we can see, we have built a counts matrix of each of the main characters and how often they are mentioned in each Chapter. As expected, Alice is present in every chapter being the protagonist. For those who have read the story, this will be in line with the characters present in each Chapter. In some cases, it is clear simply from the Chapter title - the Mad Hatter has 33 mentions in VII. A Mad Tea-Party and the Queen of Hearts has 37 mentions in VIII. The Queen’s Croquet-Ground.

The significance of this analysis lays in that we didn't need to specify the names of any main characters (outside of neat formatting for the visualisation) - these are simply the proper nouns that occurred more than 25 times throughout the text. We could generalise what we've created here to a pipeline that could be applied to any novel or text that meets a few criteria (e.g. has a small collection of regularly used proper nouns). It would easily and clearly answer a fundamental question for any piece of text: 'Who features in it and where?'.

### Sentiment Analysis

```python
pd.set_option('display.max_colwidth', None)
```

There are 3 core points of interest that we shall explore here:
- Compound Sentiment by Chapter
- Average Sentence Sentiment by Character
- How positively/negatively is Alice portrayed across the Chapters?

#### Compound Sentiment by Chapter

We shall use NLTK's (Vader) sentiment intensity analyser. It produces 4 scores for any input text: Negative, Neutral, Positive and Compound. Compound Sentiment is a normalized and weighted composite score varying from -1 (negative) to 1 (positive). 

The Vader Sentiment GitHub repo: https://github.com/cjhutto/vaderSentiment

```python
for i in range(0,12):
    print(chaps[i].ljust(37,' '),sia.polarity_scores(aiw_l[i]))
## Prints sentiment polarity scores of each Chapter.
```

The compound sentiment of each Chapter is quite heavily positive, except for V (-0.750) and IX (-0.999) where there is a stronger negativity. From the other scores, we see that Chapter V suffers from a low amount of positive words (as opposed to an abundance of negative words) whereas Chapter IX has the highest proportion of negative words and lowest proportion of positive words out of any chapter.

In these two chapters, we see that one consists of a frustrating, dry conversation between Alice and the Caterpillar. The other is a narrative describing the sad, unfortunate story of the Mock Turtle. It would make sense that these two chapters reflect an overall negative sentiment.

#### Average Sentence Sentiment by Character

We now want to create a dataframe segmenting the chapters and text by sentence. We use NLTK's sentence tokenizer, then also word_tokenize as we've done before, but this is only for sampling the data and not used in sentiment analysis.

```python
df3 = pd.DataFrame()
for i in range(0,12):
    df3 = pd.concat([df3,pd.DataFrame([[i+1,x] for x in sent_tokenize(aiw2_l[i])])])
df3.rename({0:'Chapter',1:'Sentence'},axis='columns',inplace=True)
df3 = df3[df3['Sentence']!='']
## Tokenize Sentences of each Chapter.
df3['Sent_Tokens'] = df3['Sentence'].apply(lambda l: [x.lower().replace('—','').replace('-','').replace('_','') for x in word_tokenize(l) if (x not in '.,;:_?!“”‘’"\'£$%^&*()') and (x.lower() not in stop_words)])
## Word Tokenize each individual sentnece as we've done before.
df3['Sentiment'] = df3['Sentence'].apply(lambda x: sia.polarity_scores(x)['compound'])
## Apply Sentiment scoring to each sentence.
```

```python
print('Character    Average Sentence Sentiment')
for char in list(main_char['Token'].unique()):
    print(char.ljust(12,' '),df3.loc[df3['Sent_Tokens'].apply(lambda x: char in x),'Sentiment'].mean())
## Print each main character and the average sentiment of all sentences mentioning them.
```

Unfortunately, we don't see much consistency across average sentence sentiment. The majority of sentences in the text are neutral, which largely waters down any analysis or trends. The only exception is cat, Cheshire Cat, who has particularly positive sentences. Let us take a look.

```python
print('Sentences containing both \'cat\' and \'grin\':',len(df3.loc[df3['Sent_Tokens'].apply(lambda x: ('cat' in x)) & df3['Sentence'].str.contains('grin')]))
## All sentences featuring the word 'cat' and a word stemming from 'grin'.
df3.loc[df3['Sent_Tokens'].apply(lambda x: ('cat' in x))]
## All sentences featuring the word 'cat'.
```

Those who are familiar with Alice in Wonderland will know that Cheshire Cat is well known for its regular and almost sinister grin. In fact, out of all 32 sentences including 'cat', 6 also contain a word stemming from 'grin'. This combined with the cat's polite language (e.g. a regular use of the word 'please') gives all sentences relating to Cheshire Cat a very positive sentiment.

We also must concede some flaws to our processing. Cheshire Cat is not the only cat mentioned in this story - a few sentences refer instead to Alice's cat Dinah. Alice speaks very positively of her cat, also adding extra positive sentiment to all 'cat' sentences.

#### How positively/negatively is Alice portrayed across the Chapters?

We shall purposely try to visualise this answer, so first we transform and merge our sentiment data to create a 12-row DataFrame. Each row is a Chapter, with two columns for average sentence sentiment, and average sentiment of sentences containing 'alice'.

```python
sent_chap = df3.groupby('Chapter').agg({'Sentiment':'mean'})
sent_chap = sent_chap.merge(df3[df3['Sent_Tokens'].apply(lambda x: 'alice' in x)].groupby('Chapter').agg({'Sentiment':'mean'}),how='inner',on='Chapter')
sent_chap.rename({'Sentiment_x':'Average Sentence Sentiment','Sentiment_y':'Average Alice Sentence Sentiment'},axis='columns',inplace=True)
## Aggregate sentence sentiment dataframe twice (one all, one just 'alice' sentences), inner join together and relabel.
```

```python
## Seaborn Barplot Code for Average Sentence Sentiment and Average Alice-Sentence Sentiment.
fig, ax1 = plt.subplots(figsize=(17,8))

ax1.bar([x-0.1 for x in range(0,12)],sent_chap['Average Sentence Sentiment'],
        width=0.2,label='All Sentences',color="#0099ff",zorder=2)
ax1.bar([x+0.1 for x in range(0,12)],sent_chap['Average Alice Sentence Sentiment'],
        width=0.2,label='Alice-mentioning Sentences',color="#ff9933",zorder=2)

ax1.legend(loc="upper center",fontsize="large",title="Sentence Type",title_fontsize="x-large", ncol=4)
ax1.set_xlabel("Chapter", size=16, labelpad=20)
ax1.set_ylabel("Compound Sentiment Score", size=16, labelpad=20)
ax1.set_title("Average Sentence Sentiment by Chapter", size=20, pad=20)
ax1.set_xticks([x for x in range(0,12)])
ax1.set_xticklabels(chaps,rotation=30,ha="right",rotation_mode="anchor")
ax1.axhline(0,color='black')

plt.subplots_adjust(bottom=0.1,  
                    top=0.9, 
                    wspace=0.3, 
                    hspace=0.2);
```

We can see that the sentiment of sentences containing 'Alice' are generally in line with average sentence sentiment. We will only comment on Chapter 11, where we see a differnence of >0.25 between the sentiment of Alice-sentences and all sentences.

For a literary explanation, Chapter 11 sees a serious scenario, a court of justice, and Alice's particularly frivolous reactions and casual behaviour towards the weight of the situation - given its riduculousness. This causes her dialogue to have a positive sentiment, reflecting the light-hearted commentary, whilst the other characters remain quite serious.

Our sentiment scoring comes from multiple unidentified contexts. Alice's dialogue, the narrator's descriptions of Alice and other character's dialogues speaking of Alice are three examples of disjoint cases when positive/negative language can mean different things. Whilst we have overall rated the sentiment of words around 'Alice' mentions, we haven't dissected the context they are said in. We shall not explore this further, but one could use quotation marks and dialogue verbs (e.g. 'said') to label the dialogue context (e.g. was it a quote, who said it) and enrich the potential analysis.
